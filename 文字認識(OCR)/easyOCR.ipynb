```python
# インポート
import easyocr
import cv2

from google.colab.patches import cv2_imshow
```

```python
reader = easyocr.Reader(['ja','en']) # 文字の選択 一回だけでいい
```

renderの中身　[(座標),(文字),(信頼度)]

```python
path = "youer/path"
```

```python
result = reader.readtext(path)
```

```python
result2 = reader.readtext(path, detail=0) # 文章のみ
```

```python
img = cv2.imread(path)
img_1 = img.copy()
# 文字ごとに枠を作成
for i in range(len(result)):
    cv2.rectangle(img_1, result[i][0][0], result[i][0][2], (0,255,0),3)
```

```python
# goolecolaboratory
cv2_imshow(img_1)
```
